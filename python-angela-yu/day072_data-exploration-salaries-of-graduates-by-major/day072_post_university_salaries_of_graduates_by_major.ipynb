Selanjutnya, kita amati jurusan yang punya risiko rendah. Dikatakan rendah risiko, kalau selisih mid-career salary antara persentil ke-90 dan ke-10 itu kecil.

Buat bisa ngamatin itu, kita bikin kolom baru: Spread

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col) # 1 menandakan posisi masukin kolomnya, 'Spread' buat rename nama kolom
clean_df.head()
```

Selanjutnya, kita urutkan semua nilai di kolom Spread. Jadi kebayang jurusan mana aja yang rendah risiko.

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

daaan itu dia 5 jurusan teratas yang rendah risiko.

# Challenge:
- Caritau 5 jurusan teratas dengan salary tertinggi di persentil ke-90
- Jurusan apa yang spread nya paling tinggi, a.k.a yang paling tinggi risikonya?

```python
# 5 jurusan teratas dengan salary tertinggi di persentil ke-90
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# jurusan yang spreadnya paling tinggi, alias yang paling tinggi risiko
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```
